#### Modifique o programa para trabalhar com duas filas. Para facilitar seu trabalho, considere o comando A para atendimento da fila 1; B para atendimento da fila 2. O mesmo para a chegada de clientes: F para fila 1; e G, para fila 2.

```python
ultimo_a = 10
fila_1 = list(range(1, ultimo_a + 1))

ultimo_b = 10
fila_2 = list(range(1, ultimo_b + 1))

atendimento_hj = 0

while True:
    print(f'Existem {len(fila_1)} clientes na fila 1.')
    print(f'Existem {len(fila_2)} clientes na fila 2.')
    print(f'Fila atual: {fila_1}.')
    print(f'Fila atual: {fila_2}.')
    print('Digite F para adicionar um cliente ao final da fila 1.')
    print('Digite G para adicionar um cliente ao final da fila 2.')
    print('Ou A para realizar o atendimento da fila 1. S para sair')
    print('Ou B para realizar o atendimento da fila 2. S para sair')

    operacao = input('Operação (F, G, A, B ou S): ')

    if operacao == 'A':
        if len(fila_1) > 0:
            atendido = fila_1.pop(0)
            atendimento_hj += 1
            print(f'Cliente {atendido} atendido.')
        else:
            print('Fila 1 vazia! Ninguém para atender.')
    elif operacao == 'F':
        ultimo_a += 1 # incremento do ticket do novo cliente
        fila_1.append(ultimo_a)
    
    if operacao == 'B':
        if len(fila_2) > 0:
            atendido = fila_2.pop(0)
            atendimento_hj += 1
            print(f'Cliente {atendido} atendido.')
        else:
            print('Fila 2 vazia! Ninguém para atender.')
    elif operacao == 'G':
        ultimo_b += 1 # incremento do ticket do novo cliente
        fila_2.append(ultimo_b)
    
    elif operacao == 'S':
        break
    else:
        print('Operação inválida! Digite apensas F, G, A, B ou S!')
print(f'Hoje foram atendidos {atendimento_hj} clientes.')
```
